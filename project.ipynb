```python
df['date']=pd.to_datetime(df['date'],errors='coerce')
```

```python
df.dtypes
```

```python
df['year']=df['date'].dt.year
df['month']=df['date'].dt.month_name()
df['year'] = df['year'].astype('Int64')
```

```python
df['company']=df['company'].str.strip()
```

```python
df['company'].nunique()
```

```python
df
```

```python
#Top 10 companies by layoff
df.groupby('company')['total_laid_off'].sum().sort_values(ascending=False).head(10)
```

```python
#top10 countries by layoff 
df.groupby('country')['total_laid_off'].sum().sort_values(ascending=False).head(10)
```

```python
#WHich company layoff is 100%
df[df['percentage_laid_off'] == 1]['company']
```

```python
#which yr has the highest layoff
df.groupby('year')['total_laid_off'].sum().sort_values(ascending=False).head(1)
```

```python
#which month has max layoff
df.groupby('month')['total_laid_off'].sum().sort_values(ascending=False)
```

```python
df
```

```python
#which sector is suffered most
df.groupby('industry')['total_laid_off'].sum()
```

```python
#Which companies had layoffs more than once?
count=df.groupby('company').size()
count[count> 1]
```

```python
#Industry vs Year layoffs

pd.pivot_table(df,values='total_laid_off',index='industry',columns='year',aggfunc=sum)
```

```python
#layoff by company stage
df.groupby('stage')['total_laid_off'].sum().sort_values(ascending=False)
```

```python
#Find companies where layoffs happened more than once AND total layoffs are very high
count=df.groupby('company').size()
total=df.groupby('company')['total_laid_off'].sum()
result=(count>1) & (total > 1000)
df.groupby('company')['total_laid_off'].sum()[result]
```
